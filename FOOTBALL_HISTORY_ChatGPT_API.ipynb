```python
def update_chat(messages, role, content):
  messages.append({"role": role, "content": content})
  return messages
```

```python
def get_chatgpt_response(messages):
  response = openai.ChatCompletion.create(
  model="gpt-3.5-turbo",
  messages=messages
)
  return  response['choices'][0]['message']['content']
```

```python
import pprint

messages=[
   {"role": "system", "content": "You are a helpful AI Football History Expert."},
        {"role": "user", "content": "I am Akash, I want to learn about football history"},
        {"role": "assistant", "content": "Hello, Akash Thats awesome, what do you want to know aboout football history"}
  ]

while True:
  pprint.pprint(messages)  # brings the initial conversation to the point we already had
  user_input = input()     # enabling chat conversation
  messages = update_chat(messages, "user", user_input)   # appending record using the helper function we defined
  model_response = get_chatgpt_response(messages)
  messages = update_chat(messages, "assistant", model_response)
```
